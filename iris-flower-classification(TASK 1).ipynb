# USING TRAIN_TEST_SPLIT

```python
from sklearn.model_selection import train_test_split
```

```python
df_train,df_test = train_test_split(df,test_size = 0.25)
```

# PREPARE DATA FOR MODELLING

```python
x_train = df_train.drop(columns = ['Species','species_name']).values
y_train = df_train["Species"].values
```

# FINDING ACCURACY MANUALLY

```python
def single_feature_prediction(petal_length):
    """Predicts the Iris species given the petal length"""
    if petal_length < 2.5:
        return 0
    elif petal_length<4.8:
        return 1
    else: 
        return 2
```

```python
manual_y_predictions = np.array([single_feature_prediction(val) for val in x_train[:, 2]])
```

```python
manual_model_accuracy = np.mean(manual_y_predictions == y_train)
```

```python
print(f'Manual model accuracy  : {manual_model_accuracy *100:.2f}%')
```

### Got Accuracy of 95% manually

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import Lasso
from sklearn.svm import SVC
from sklearn.linear_model import Ridge
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=cross_val_score(KNeighborsClassifier(n_neighbors=11), xt, yt,cv=8)
np.average(knn)
```

```python
logistic=cross_val_score(LogisticRegression(), xt, yt,cv=8)
np.average(logistic)
```

```python
ridge=cross_val_score(Ridge(), xt, yt,cv=8)
np.average(ridge)
```

```python
svm=cross_val_score(SVC(), xt, yt,cv=8)
np.average(svm)
```

```python
linear=cross_val_score(LinearRegression(), xt, yt,cv=8)
np.average(linear)
```

## From above we can see the maximum accuracy is given by Logistic regression So we will use this algorithm
